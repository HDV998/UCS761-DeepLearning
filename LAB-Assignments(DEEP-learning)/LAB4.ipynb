# Multiple Linear Regression from Scratch

Using only NumPy, we build a linear regression model trained via gradient descent. The dataset contains age and years of experience as predictors for income.

```python
import pandas as pd
import numpy as np
salary_df = pd.read_csv("multiple_linear_regression_dataset.csv")
print("First 5 rows:\n", salary_df.head().to_string())
print("\nColumns:", salary_df.columns.tolist())
print("Shape:", salary_df.shape)
```

## Separating Features and Target

Age and experience are our two input features. Income is what we aim to predict.

```python
X = salary_df[["age", "experience"]].values
y = salary_df["income"].values
print("Feature matrix X shape:", X.shape)
print("Target vector y shape: ", y.shape)
```

## Weight Initialization

One weight per input feature. Both weights and bias start at zero.

```python
n_features = X.shape[1]
W = np.zeros(n_features)
b = 0.0
print(f"Initialized W = {W}, b = {b}")
```

## Linear Prediction

The model computes $\hat{y} = X \cdot W + b$. No activation — we need unbounded real output.

```python
def predict(X, W, b):
    return X.dot(W) + b

sample_pred = predict(X[:3], W, b)
print("Initial predictions (first 3):", sample_pred)
```

## Mean Squared Error

MSE squares residuals so large errors are penalized much more than small ones.

```python
def mse_loss(y_true, y_pred):
    residuals = y_pred - y_true
    return (residuals ** 2).mean()

init_loss = mse_loss(y, predict(X, W, b))
print(f"Initial MSE Loss: {init_loss:.2f}")
```

## Gradients and Update Step

```python
def compute_gradients(X, y_true, y_pred):
    n = len(y_true)
    diff = y_pred - y_true
    grad_W = (2.0 / n) * X.T.dot(diff)
    grad_b = (2.0 / n) * diff.sum()
    return grad_W, grad_b

def gradient_descent_step(W, b, gW, gb, lr):
    W = W - lr * gW
    b = b - lr * gb
    return W, b
```

## Training Loop

1000 epochs, small learning rate due to large income values.

```python
alpha = 0.0001
epochs = 1000
for i in range(epochs):
    y_pred = predict(X, W, b)
    loss   = mse_loss(y, y_pred)
    gW, gb = compute_gradients(X, y, y_pred)
    W, b   = gradient_descent_step(W, b, gW, gb, alpha)
    if i % 100 == 0:
        print(f"Epoch {i:4d}  |  MSE Loss: {loss:.2f}")
```

## Results and Prediction

```python
print(f"Learned weights (age, experience): {np.round(W,4)}")
print(f"Learned bias: {b:.4f}")
new_sample = np.array([50, 15])
income_est = new_sample.dot(W) + b
print(f"Estimated income for age=50, exp=15: {income_est:.2f}")
```
